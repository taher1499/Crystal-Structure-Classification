```python
df=df.drop(['Formation energy [eV/atom]','Stability [eV/atom]','Magnetic moment [mu_B]','Volume per atom [A^3/atom]','a [ang]','b [ang]','c [ang]','alpha [deg]','beta [deg]','gamma [deg]','Vacancy energy [eV/O atom]','Band gap [eV]'],axis=1)
```

```python
df['Goldschmidt Tolerance']=(df['Radius A [ang]']+1.4)/((2**.5)*(df['Radius B [ang]']+1.4))
```

```python
df['Octahedral factor']=(df['Radius B [ang]']/1.4)
```

```python
new_data=new_data[['ABO3','rA at 12 [ang]','rA at 6 [ang]','rB at 6 [ang]','VA/VB','tR']]
```

```python
df=pd.merge(df,new_data,how='left',left_on='Chemical formula',right_on='ABO3')
```

```python
df=pd.merge(df,en_bl,how='left',left_on='Chemical formula',right_on='ABO3')
```

```python
df['New Tolerance Factor']=((df['rA at 6 [ang]']/df['rA at 12 [ang]'])/3)*((df['Radius A [ang]']+1.4)/(df['Radius B [ang]']+1.4))*((df['Radius B [ang]']+1.4)/1.4*(2**0.5))*((df['Radius A [ang]']+1.4)/1.4*(2**0.5))
```

```python
df['EN diff']=(df['EN of B']+df['EN of A']-6.88)/2
```
